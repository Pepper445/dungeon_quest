```python
import random
from IPython.display import clear_output  # Clears notebook output

def print_summary(player_name, player_health, player_inventory, player_quit, player_dead):
    clear_output(wait=True)
    print(f"\nGame Over Summary for {player_name}:")
    print(f"Final Health: {player_health}")
    print(f"Final Inventory: {player_inventory}")
    total_value = sum(treasure_dict.get(item, 0) for item in player_inventory)
    print(f"Total Treasure Value: {total_value} gold")
    if player_quit and not player_dead:
        print("\nYou chose to quit the game early. Thanks for playing!")
    elif player_dead:
        print("\nYou died in the dungeon. Better luck next time!")
    else:
        print("\nCongratulations! You made it through all 5 rooms!")

# Step 1: Player Setup
player_name = input("What shall we call you?: ")
player_health = 10
player_inventory = []

# Step 2: Treasure Dictionary
treasure_dict = {
    "Gold Coin": 10,
    "Silver Coin": 5,
    "Gemstone": 50,
    "Ancient Artifact": 100
}

# Flags to track quitting or death
player_quit = False
player_dead = False

# Step 3: Game Loop - Moving through 5 rooms
room = 1
while room <= 5:
    if not player_dead and not player_quit:
        clear_output(wait=True)  # Clear previous output
        print(f"{player_name}, you are in Room {room}!")
        print("\nOptions:")
        print("1. Search for treasure")
        print("2. Move to next room")
        print("3. Check health and inventory")
        print("4. Quit the game")

    choice = input("Choose an option (1-4): ")

    if choice == "1":
        event = random.choice(["treasure", "trap"])
        if event == "treasure":
            treasure_name = random.choice(list(treasure_dict.keys()))
            player_inventory.append(treasure_name)
            print(f"\nYou found a {treasure_name} worth {treasure_dict[treasure_name]} gold!")
        else:
            damage = random.randint(1, 3)
            player_health -= damage
            print(f"\nOh no! You triggered a trap and lost {damage} health points!")

        if player_health <= 0:
            player_dead = True
            print("\nWhomp, whomp, whomp. X_X You have died! Game Over.")
            input("\nPress Enter to see your summary...")
            print_summary(player_name, player_health, player_inventory, player_quit, player_dead)
            break  # Exit loop if dead

        input("\nPress Enter to continue...")

    elif choice == "2":
        print("\nMoving to the next room...")
        input("\nPress Enter to continue...")
        room += 1  # Go to next room

    elif choice == "3":
        print(f"\nHealth: {player_health}")
        print(f"Inventory: {player_inventory}")
        input("\nPress Enter to continue...")

    elif choice == "4":
        print("\nYou have chosen to quit the game.")
        player_quit = True
        input("\nPress Enter to see your summary...")
        print_summary(player_name, player_health, player_inventory, player_quit, player_dead)
        break  # Exit the loop to go to summary

    else:
        print("\nInvalid choice. Please enter 1-4.")
        input("\nPress Enter to continue...")

# If the player finishes all rooms without quitting or dying, show summary
if not player_quit and not player_dead:
    print_summary(player_name, player_health, player_inventory, player_quit, player_dead)
```
